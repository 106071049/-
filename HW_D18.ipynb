```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-90,270)
y = np.cos(x*np.pi/180)
```

```python
plt.plot(x,y)
plt.show()
```

```python
n = 1024
w = np.random.normal(0,1,n)
z = np.random.normal(0,1,n)

tan = np.arctan2(z,w)
plt.scatter(w,z, s=50, c=tan, alpha=10)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
